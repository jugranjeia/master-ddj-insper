**METODOLOGIA**
 
O cálculo do preço do voto foi feito a partir dos dados disponibilizados pelo TSE (Tribunal Superior Eleitoral) e com uso da biblioteca Pandas para análise de dados com Python. Primeiramente, analisamos os gastos dos candidatos ao governo do estado do Rio de Janeiro durante a campanha de 2022 e cruzamos com o total de votos.

Castro gastou R 19.322.516,67 e obteve 4.930.288 votos (58,6%) . Dividindo os gastos totais pela quantidade de votos, chegamos ao custo de cada voto individual em Castro - R$ 3,92.

É preciso considerar que quanto maior a quantidade de votos que um candidato recebe, mais diluído será o valor gasto durante sua campanha. Logo, o candidato eleito terá um valor de voto menor.
