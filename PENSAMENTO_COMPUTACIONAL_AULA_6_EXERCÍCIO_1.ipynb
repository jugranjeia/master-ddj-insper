```python
import nltk
import re
nltk.download('punkt')
```

```python
with open('esau_e_jaco.txt', 'r') as arquivo:
    livro_completo = arquivo.read().replace('\n', '').lower().replace(':', '').replace(',','').replace('.','').replace('!','').replace('?','').replace(';','').replace('—','').replace('``','')
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

```python
texto = livro_completo
texto_tokens = word_tokenize(texto)

texto_sem_stopwords = [palavra for palavra in texto_tokens if not palavra in stopwords.words()]

print(texto_sem_stopwords)
```

```python
contador_palavras = {}
for frase in texto_sem_stopwords:
    palavras = nltk.word_tokenize(frase)
    for palavra in palavras:
        if palavra not in contador_palavras.keys():
            contador_palavras[palavra] = 1
        else:
            contador_palavras[palavra] += 1
            
print (contador_palavras)
```

```python
palavras_repetidas = dict(sorted(contador_palavras.items(), key=lambda item: item[1], reverse=True))
palavras_repetidas
```

```python
top5 = list(palavras_repetidas)[:5]
```

```python
f"As cinco palavras mais usadas no livro Esaú e Jacó são {top5}"
```

Crie um código que receba o PDF de um livro e uma palavra e diga qual quantas vezes aquela
palavra é usada

```python
def analise_livro(pdf):
  livro2 = PdfReader(pdf)
  numero_paginas2 = len(livro2.pages)

  with open((pdf),'rb') as arquivo_pdf2, open('novo.txt', 'w') as arquivo_txt2:
    ler_pdf2 = PyPDF2.PdfFileReader(arquivo_pdf2)
    numero_de_paginas2 = ler_pdf2.getNumPages()
    for numero_de_paginas2 in range(0,int(numero_paginas2)):
        pagina2 = ler_pdf2.getPage(numero_de_paginas2)
        conteudo_pagina2 = pagina2.extractText()
        arquivo_txt2.write(conteudo_pagina2)

  with open('novo.txt', 'r') as arquivo2:
    livro_completo2 = arquivo2.read().replace('\n', '').lower().replace(':', '').replace(',','').replace('.','').replace('!','').replace('?','').replace(';','').replace('—','').replace('``','')

  texto2 = livro_completo2
  texto_tokens2 = word_tokenize(texto2)
  texto_sem_stopwords2 = [palavra for palavra in texto_tokens2 if not palavra in stopwords.words()]

  contador_palavras2 = {}
  for frase in texto_sem_stopwords2:
    palavras2 = nltk.word_tokenize(frase)
    for palavra in palavras2:
      if palavra not in contador_palavras2.keys():
        contador_palavras2[palavra] = 1
      else:
        contador_palavras2[palavra] += 1

  palavras_repetidas2 = dict(sorted(contador_palavras2.items(), key=lambda item: item[1], reverse=True))
  novo_top5 = list(palavras_repetidas2)[:5]

  print(f"As cinco palavras mais usadas no livro são {novo_top5}")
  
  num_chave = (palavras_repetidas2[chave])
  if chave in palavras_repetidas2:
    print(f'A palavra {chave} está no livro Iaiá e aparece {num_chave} vezes!')
  else:
    print(f'A palavra {chave} não está presente no livro Iaiá. Tente novamente')
```
